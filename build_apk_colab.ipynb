## 第 2 步：上传项目源码

**点击左侧文件夹图标** → 将 `ctyun-quote-colab.zip` 拖入文件区域

等待上传完成后，运行下方代码解压。

```python
import zipfile, os

# 检查上传的文件
if not os.path.exists('ctyun-quote-colab.zip'):
    print("❌ 未找到 ctyun-quote-colab.zip，请先上传！")
    print("   👉 点击左侧文件夹图标，将 zip 文件拖入")
else:
    # 解压到项目目录
    with zipfile.ZipFile('ctyun-quote-colab.zip', 'r') as z:
        z.extractall('ctyun-quote')
    os.chdir('ctyun-quote')
    print(f"✅ 解压完成！项目目录: {os.getcwd()}")
    print(f"   ├── app.py（Flask 服务端）")
    print(f"   ├── android_main.py（Android 入口）")
    print(f"   ├── templates/index.html（前端页面）")
    print(f"   └── buildozer.spec（构建配置）")
```

## 第 3 步：编译 APK

```python
# 开始编译
!buildozer android debug

print("\n✅ 编译完成！")
```

## 第 4 步：下载 APK

编译完成后，运行下方代码将 APK 复制到工作目录，然后手动下载。

```python
import glob, shutil

# 查找生成的 APK
apk_files = glob.glob('bin/*.apk') + glob.glob('**/bin/*.apk', recursive=True)

if apk_files:
    for apk in apk_files:
        # 复制到工作目录便于下载
        dest = f'/content/{os.path.basename(apk)}'
        shutil.copy(apk, dest)
        size_mb = os.path.getsize(apk) / (1024*1024)
        print(f"📱 APK 文件: {os.path.basename(apk)}")
        print(f"   大小: {size_mb:.1f} MB")
        print(f"   路径: {dest}")
    print("\n👉 在左侧文件列表找到 APK 文件，右键 → 下载")
else:
    print("❌ 未找到 APK 文件，编译可能失败了")
```
